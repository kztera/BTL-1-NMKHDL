# Elasticsearch  Search Queries , Filters & Aggregations

## Initiate modules

```python
from elasticsearch import Elasticsearch
```

## Initiate global variables

```python
client = Elasticsearch()
indexName = "medical"
docType="diseases"
searchFrom = 0
searchSize= 3
```

## Finding Lupus

### Finding Lupus first trial

```python
searchBody={
"fields":["name"],
"query":{
    "simple_query_string" : {
        "query": '+fatigue+fever+"joint pain"',
        "fields": ["fulltext","title^5","name^10"]
        }
    }
}
client.search(index=indexName,doc_type=docType, body=searchBody, from_ = searchFrom, size=searchSize)
```

### Finding Lupus second trial

```python
searchBody={
"fields":["name"],
"query":{
    "simple_query_string" : {
        "query": '+fatigue+fever+"joint pain"+rash',
        "fields": ["fulltext","title^5","name^10"]
        }
    }
}
client.search(index=indexName,doc_type=docType, body=searchBody, from_ = searchFrom, size=searchSize)
```

### Finding Lupus third trial

```python
searchBody={
"fields":["name"],
"query":{
    "simple_query_string" : {
        "query": '+fatigue+fever+"joint pain"+rash+"chest pain"',
        "fields": ["fulltext","title^5","name^10"]
        }
    }
}
client.search(index=indexName,doc_type=docType, body=searchBody, from_ = searchFrom, size=searchSize)
```

### Fuzzy search example

```python
searchBody={
"fields":["name"],
"query":{
    "simple_query_string" : {
        "query": 'lupsu~2  ',
        "fields": ["name^10"],
        }
    }
}
client.search(index=indexName,doc_type=docType, body=searchBody, from_ = searchFrom, size=searchSize)
```

## Finding Heart Attack

```python
searchBody={
"fields":["name"],
"query":{
    "simple_query_string" : {
        "query": '+"left arm pain"~10',
        "fields": ["fulltext","title^5","name^10"]
        }
    },
"highlight" : {
        "fields" : {
            "fulltext" : {}
        }
    }
}
client.search(index=indexName,doc_type=docType, body=searchBody, from_ = searchFrom, size=searchSize)
```

## Filters & Aggregations

### Diabetes KeyWords

```python
searchBody={
"fields":["name"],
"query":{
    "filtered" : {
        "filter": {
            'term': {'name':'diabetes'}
        }
    }
},  
"aggregations" : {
        "DiseaseKeywords" : {
            "significant_terms" : { "field" : "fulltext", "size" : 30 }
        }
    }
}
client.search(index=indexName,doc_type=docType, body=searchBody, from_ = searchFrom, size=searchSize)
```

### Advanced Index

return here after creating the diseases2 doctype

```python
client = Elasticsearch()
indexName = "medical"
docType="diseases2"
searchFrom = 0
searchSize= 3
```

```python
searchBody={
"fields":["name"],
"query":{
    "filtered" : {
        "filter": {
            'term': {'name':'diabetes'}
        }
    }
},  
"aggregations" : {
        "DiseaseKeywords" : {
            "significant_terms" : { "field" : "fulltext", "size" : 30 }
        },
        "DiseaseBigrams": {
            "significant_terms" : { "field" : "fulltext.shingles", "size" : 30 }
        }
    }
}
client.search(index=indexName,doc_type=docType, body=searchBody, from_ = searchFrom, size=searchSize)
```
